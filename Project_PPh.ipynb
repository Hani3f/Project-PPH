## Data exploration

### Subtask:
Explore the loaded data to understand its structure, identify data types, check for missing values, and examine the statistical distribution of numerical features.

## Data cleaning

### Subtask:
Clean the data by handling missing values and inconsistencies.


**Reasoning**:
Clean the data by handling missing values, inconsistencies, and duplicates, then verify the result.



```python
# Convert relevant columns to numeric, handling errors
for col in ['persediaan_awal', 'persediaan_akhir', 'hpp_rawat_inap', 'hpp_rawat_jalan', 'total_hpp',
            'pendapatan_operasional', 'pendapatan_penjualan', 'ppn_keluaran', 'ppn_masukan', 'ppn_netto',
            'laba_kotor_penjualan', 'beban_operasional', 'overhead_farmasi', 'penyusutan_total',
            'penyusutan_garis_lurus', 'penyusutan_saldo_menurun', 'penyusutan_fiskal',
            'laba_kotor_operasional', 'laba_sebelum_pajak', 'laba_kena_pajak', 'pph_badan_normal',
            'tarif_pph_tax_holiday', 'pph_badan_tax_holiday', 'laba_bersih_normal',
            'laba_bersih_tax_holiday', 'kerugian_fiskal']:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False), errors='coerce')

# Remove columns with more than 50% missing values
missing_percentage = df.isnull().sum() / len(df) * 100
columns_to_drop = missing_percentage[missing_percentage > 50].index
df = df.drop(columns=columns_to_drop)

# Impute missing values for remaining numerical columns using the median
numerical_cols = df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].median())

# Remove duplicate rows
df = df.drop_duplicates()

# Verify the cleaning
print(df.info())
missing_values_after = df.isnull().sum()
print("\nMissing Values After Cleaning:\n", missing_values_after)

display(df.head())
```

## Data wrangling

### Subtask:
Data Wrangling and Feature Engineering


**Reasoning**:
Group the data by year and calculate the sum of relevant financial metrics. Then, calculate the year-over-year growth rates for these metrics, handling potential zero values. Finally, engineer additional features like profit margin.



```python
# Group data by year and sum relevant metrics
df_yearly = df.groupby('tahun').sum()

# Calculate year-over-year growth rates
for col in ['pendapatan_operasional', 'pendapatan_penjualan', 'laba_kotor_penjualan', 'beban_operasional']:
    df_yearly[f'{col}_growth'] = (df_yearly[col] - df_yearly[col].shift(1)) / df_yearly[col].shift(1)
    df_yearly[f'{col}_growth'] = df_yearly[f'{col}_growth'].fillna(0)  # Handle the first year (no previous year)

# Calculate profit margin
df_yearly['profit_margin'] = (df_yearly['laba_kotor_penjualan'] / df_yearly['pendapatan_penjualan'])
df_yearly['profit_margin'] = df_yearly['profit_margin'].fillna(0)

display(df_yearly)
```

```python
# Calculate year-over-year changes
for col in ['total_hpp', 'pendapatan_operasional', 'laba_sebelum_pajak', 'laba_bersih_normal']:
    df[f'yoy_{col}'] = df.groupby('tahun')[col].pct_change() * 100
    df[f'yoy_{col}'].fillna(0, inplace=True)  # Handle the first year's data

# Create financial ratios
df['gross_profit_margin'] = (df['pendapatan_operasional'] - df['total_hpp']) / df['pendapatan_operasional']
df['tax_burden'] = df['laba_kena_pajak'] / df['laba_sebelum_pajak']
df['net_profit_margin'] = df['laba_bersih_normal'] / df['pendapatan_operasional']
# Handle potential division-by-zero errors
for col in ['gross_profit_margin', 'tax_burden', 'net_profit_margin']:
    df[col].fillna(0, inplace=True)

# Calculate tax holiday impact
df['tax_holiday_impact'] = df['laba_bersih_normal'] - df['laba_bersih_tax_holiday']

display(df.head())
```

**Reasoning**:

```python
# Calculate year-over-year changes
for col in ['total_hpp', 'pendapatan_operasional', 'laba_sebelum_pajak', 'laba_bersih_normal']:
    df[f'yoy_{col}'] = df.groupby('tahun')[col].pct_change(fill_method=None) * 100
    df[f'yoy_{col}'] = df[f'yoy_{col}'].fillna(0)

# Create financial ratios
df['gross_profit_margin'] = (df['pendapatan_operasional'] - df['total_hpp']) / df['pendapatan_operasional']
df['tax_burden'] = df['laba_kena_pajak'] / df['laba_sebelum_pajak']
df['net_profit_margin'] = df['laba_bersih_normal'] / df['pendapatan_operasional']

# Handle potential division-by-zero errors
for col in ['gross_profit_margin', 'tax_burden', 'net_profit_margin']:
    df[col] = df[col].fillna(0)

# Calculate tax holiday impact
df['tax_holiday_impact'] = df['laba_bersih_normal'] - df['laba_bersih_tax_holiday']

display(df.head())
```
